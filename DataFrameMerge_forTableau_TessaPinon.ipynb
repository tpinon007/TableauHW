```python
# Dependencies

import pandas as pd
import numpy as np
```

```python
#Read in modified csv of Dec. 2017 data
dec17_df=pd.read_csv("../TableauHW/201712-citibike-tripdata_mod.csv")
dec17_df.head()
```

```python
#Modify column names in Dec. 2017 data frame
moddec17_df=dec17_df.rename(columns={
    'tripduration':'dec17_tripduration',
    'starttime':'dec17_starttime',
    'stopttime':'dec17_stoptime',
    'start station id':'dec17_startstationid',
    'usertype':'dec17_usertype',
    'birthyear':'dec17_birthyear',
    'gender':'dec17_gender',
})

moddec17_df.head()
```

```python
#Read in modified csv of Jan. 2018 data
jan18_df=pd.read_csv("../TableauHW/201801-citibike-tripdata_mod.csv")
jan18_df.head()
```

```python
#Modify column names in Jan. 2018 data frame
modjan18_df=jan18_df.rename(columns={
    'tripduration':'jan18_tripduration',
    'starttime':'jan18_starttime',
    'stopttime':'jan18_stoptime',
    'start station id':'jan18_startstationid',
    'usertype':'jan18_usertype',
    'birthyear':'jan18_birthyear',
    'gender':'jan18_gender',
})

modjan18_df.head()
```

```python
#Read in modified csv of Feb. 2018 data
feb18_df=pd.read_csv("../TableauHW/201802-citibike-tripdata_mod.csv")
feb18_df.head()
```

```python
#Modify column names in Jan. 2018 data frame
modfeb18_df=feb18_df.rename(columns={
    'tripduration':'feb18_tripduration',
    'starttime':'feb18_starttime',
    'stopttime':'feb18_stoptime',
    'start station id':'feb18_startstationid',
    'usertype':'feb18_usertype',
    'birthyear':'feb18_birthyear',
    'gender':'feb18_gender',
})

modfeb18_df.head()
```

```python
#Merge Jan and Feb data frames (outer join)
janfeb_df=modjan18_df.merge(modfeb18_df,left_on='jan18_tripduration',right_on='feb18_gender',how='outer')
#Merge Dec. 2017 data frame with Jan and Feb (outer join)
allwinter_df=moddec17_df.merge(janfeb_df,left_on='dec17_tripduration',right_on='feb18_gender',how='outer')
allwinter_df.head()
```

```python
janfeb_df.to_csv("janfebdata.csv", encoding='utf-8')
allwinter_df.to_csv("winterdata.csv", encoding='utf-8')
```
